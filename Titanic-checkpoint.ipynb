```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
#导入数据
train = pd.read_csv('train.csv')
train.head(10)
```

```python
test = pd.read_csv('test.csv')
test.head()
```

```python
#数据合并
titanic = pd.concat([train,test], axis=0)
titanic.head(10)
```

```python
#删除数据
del titanic['Cabin']
del titanic['Name']
del titanic['Ticket']
```

```python
titanic['Embarked'].value_counts()
```

```python
titanic['Sex'].value_counts()
```

```python
#数字型编码
col_dict = {}
cols = ['Embarked','Sex']
col_dict ={'Embarked':{'S':1,'C':2,'Q':3},'Sex':{'male':1,'female':2}}
for col in cols:
    titanic[col] = titanic[col].map(col_dict[col])
```

```python
#将数据分为训练集和测试集
from sklearn import cross_validation
y = titanic['Survived']
del titanic['Survived']
X = titanic
titanic_train, titanic_test, titanic_train_labels, titanic_test_labels \
= cross_validation.train_test_split(X, y, test_size=0.3, random_state=0)
```

```python
print titanic_train_labels.value_counts()/len(titanic_train_labels)
print titanic_test_labels.value_counts()/len(titanic_test_labels)
```

```python
#KNN算法
from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics
k_list = [1,6,9,15,21,30]
for k in k_list:
    knn_model = KNeighborsClassifier(n_neighbors = k)
    knn_model.fit(titanic_train, titanic_train_labels)
    titanic_test_pred = knn_model.predict(titanic_test)
    accuracy = metrics.accuracy_score(titanic_test_labels, titanic_test_pred)
    
    print 'k=',k
    print metrics.classification_report(titanic_test_labels,titanic_test_pred)
    print metrics.confusion_matrix(titanic_test_labels, titanic_test_pred)
    print '正确率：','%.2f'%(accuracy),'\n'
```

```python
#SVM方法
from sklearn.svm import SVC
from sklearn import metrics
kernels = ['linear','rbf','poly','sigmoid']
for kernel in kernels:
    svm_model = SVC(C = 1, kernel = kernel)
    svm_model.fit(titanic_train, titanic_train_labels)
    titanic_test_pred = svm_model.predict(titanic_test)
    accuracy = metrics.accuracy_score(titanic_test_labels, titanic_test_pred)
    
    print 'kernel=',kernel
    print metrics.classification_report(titanic_test_labels,titanic_test_pred)
    print metrics.confusion_matrix(titanic_test_labels, titanic_test_pred)
    print '正确率：','%.4f'%(accuracy),'\n'
```

```python
C_list = [0.01, 0.1, 1, 10, 100]
for C in C_list:
    svm_model = SVC(C = C, kernel = 'linear')
    svm_model.fit(titanic_train, titanic_train_labels)
    titanic_test_pred = svm_model.predict(titanic_test)
    accuracy = metrics.accuracy_score(titanic_test_labels, titanic_test_pred)
    
    print 'C=',C
    print metrics.classification_report(titanic_test_labels,titanic_test_pred)
    print metrics.confusion_matrix(titanic_test_labels, titanic_test_pred)
    print '正确率：','%.4f'%(accuracy),'\n'
```

```python
#朴素贝叶斯方法
from sklearn.naive_bayes import MultinomialNB
naive_model = MultinomialNB()
naive_model.fit(titanic_train, titanic_train_labels)
titanic_test_pred = naive_model.predict(titanic_test)
print metrics.classification_report(titanic_test_labels,titanic_test_pred)
print metrics.confusion_matrix(titanic_test_labels, titanic_test_pred)
print metrics.accuracy_score(titanic_test_labels, titanic_test_pred)
```

```python
#逻辑回归
solver_list = ['liblinear', 'sag', 'lbfgs', 'newton-cg']
for solver in solver_list:
    log_model = LogisticRegression(C=1000, solver = solver)
    log_model.fit(titanic_train, titanic_train_labels)
    titanic_test_pred = log_model.predict(titanic_test)
    
    print 'solver:',solver
    print metrics.classification_report(titanic_test_labels,titanic_test_pred)
    print metrics.confusion_matrix(titanic_test_labels, titanic_test_pred)
    print '正确率：','%.4f'%(accuracy),'\n'
```

```python
#决策树方法
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
tree_model = DecisionTreeClassifier(min_samples_leaf=8)
tree_model.fit(titanic_train,titanic_train_labels)
titanic_test_pred = tree_model.predict(titanic_test)
print metrics.classification_report(titanic_test_labels,titanic_test_pred)
print metrics.confusion_matrix(titanic_test_labels, titanic_test_pred)
print metrics.accuracy_score(titanic_test_labels, titanic_test_pred)
```

```python
#随机森林
from sklearn.ensemble import RandomForestClassifier
forest_model = RandomForestClassifier(n_estimators=100)
forest_model.fit(titanic_train, titanic_train_labels)
titanic_test_pred = forest_model.predict(titanic_test)
print metrics.classification_report(titanic_test_labels,titanic_test_pred)
print metrics.confusion_matrix(titanic_test_labels, titanic_test_pred)
print metrics.accuracy_score(titanic_test_labels, titanic_test_pred)
```
